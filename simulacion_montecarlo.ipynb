```python
import pandas as pd
```

```python
# Importamos el sales df
df_sales = pd.read_csv("sales.csv")
df_sales
```

```python
# Importamos el flights df
df_flights = pd.read_csv("flights.csv")
df_flights.shape
```

```python
# Convertir la columna 'STD' a formato datetime si aún no está en ese formato
df_flights['STD'] = pd.to_datetime(df_flights['STD'])

# Filtrar las filas donde el año de 'STD' no es 2024
df_filtered = df_flights[df_flights['STD'].dt.year != 2024]

# Eliminar filas donde la columna 'Aeronave' tiene valores nulos en el DataFrame filtrado
df_clean_flights = df_filtered.dropna(subset=['Aeronave'])

# Ajustar valores de 'Passengers' para que no excedan los de 'Capacity'
df_clean_flights['Passengers'] = df_clean_flights.apply(
    lambda row: min(row['Passengers'], row['Capacity']) if pd.notnull(row['Capacity']) else row['Passengers'],
    axis=1
)
```

```python
# Veces que se repiten los Flight_ID's
unique_Ids= df_clean_flights["Flight_ID"].value_counts()
unique_Ids
```

```python
# Definir las palabras clave a buscar en 'ProductName'
palabras_clave = ['Transfer', 'Taxi', 'Combo', 'Crew', "Vivabus", "Zona", "Vivaplay", "Apto", "Akumal", "Tulum", "Promo","Hoteles","Playa", "Autobuses", "Cancun","Hsbc-Viva","Central","Cat","Hazme"]

# Crear una expresión regular que cubra todas las palabras clave
regex = '|'.join(palabras_clave)  # Esto crea un patrón 'Transfer|Taxi|Combo|Crew|...'

# Filtrar el DataFrame para excluir filas donde 'ProductName' contiene alguna de las palabras clave
# y 'TotalSales' es negativo
df_filtrado_sales = df_sales[
    ~df_sales['ProductName'].str.contains(regex, case=False, na=False) &
    (df_sales['TotalSales'] > 0)
]
df_filtrado_sales
```

```python
# Productos unicos en df_sales
df_filtrado_sales["ProductName"].unique()
```

```python
# Juntando escalas tomandolas como un mismo vuelo y agarrando max numero de Passengers:

columnas_adicionales = [col for col in df_clean_flights.columns if col not in ['Passengers', 'Flight_ID']]  # Asegúrate de excluir 'Passengers' y 'Flight_ID'

# Agrupar por 'Flight_ID', sumar 'Passengers' y mantener el primer valor encontrado para las otras columnas
result_flights = df_clean_flights.groupby('Flight_ID', as_index=False).agg({  # `as_index=False` mantiene 'Flight_ID' como una columna y no como índice
    'Passengers': 'max',  # Obtener el valor máximo en 'Passengers'
    **{col: 'first' for col in columnas_adicionales}  # Mantener el primer valor para las otras columnas
})

result_flights
```

## Combinando datasets




```python
result_df = df_filtrado_sales.merge(result_flights[["Flight_ID", 'Capacity', 'Passengers', 'STD', 'STA']], on='Flight_ID', how='left')
result_df
```

```python
result_df['SalesProbPerProd'] = result_df['Quantity'] / result_df['Passengers'].replace(0, pd.NA)

# Mostrar el DataFrame actualizado
result_df
```

```python
flight_id_rep = result_df["Flight_ID"].value_counts()
flight_id_rep
```

```python
def filtrar_y_agregar_productos(df, flight_id, productos_esperados):
    # Filtrar el DataFrame por el Flight_ID específico
    df_filtrado = df[df['Flight_ID'] == flight_id][['Flight_ID', 'ProductType', 'ProductName', 'Quantity', "Capacity", "Passengers", 'SalesProbPerProd']]

    # Obtener un mapeo global de ProductName a ProductType basado en todo el DataFrame
    mapeo_product_type_global = df.set_index('ProductName')['ProductType'].to_dict()

    # Obtener los productos actuales bajo el Flight_ID
    productos_actuales = set(df_filtrado['ProductName'])

    # Valores representativos para capacidad y pasajeros
    capacidad_representativa = df_filtrado['Capacity'].max()
    pasajeros_representativos = df_filtrado['Passengers'].max()

    # Identificar los productos faltantes
    productos_faltantes = [prod for prod in productos_esperados if prod not in productos_actuales]

    # Crear filas nuevas para los productos faltantes y añadirlos al DataFrame filtrado
    nuevas_filas = pd.DataFrame({
        'Flight_ID': flight_id,
        'ProductType': [mapeo_product_type_global[prod] for prod in productos_faltantes],  # Usando el mapeo global
        'ProductName': productos_faltantes,
        'Quantity': [0] * len(productos_faltantes),
        'Capacity': [capacidad_representativa] * len(productos_faltantes),
        'Passengers': [pasajeros_representativos] * len(productos_faltantes),
        'SalesProbPerProd': [0.000001] * len(productos_faltantes)
    })

    # Concatenar las nuevas filas con el DataFrame filtrado si hay productos faltantes
    if not nuevas_filas.empty:
        df_filtrado = pd.concat([df_filtrado, nuevas_filas], ignore_index=True)

    return df_filtrado


productos_esperados = df_filtrado_sales["ProductName"].unique()
flight_id_especifico = '14069ac4ad5e26456a8f103484b84774'
df_actualizado = filtrar_y_agregar_productos(result_df, flight_id_especifico, productos_esperados)
df_actualizado['CumulativeProbability'] = df_actualizado['SalesProbPerProd'].cumsum()
df_actualizado
```

```python
df_actualizado['CumulativeProbability'] = df_actualizado['SalesProbPerProd'].cumsum()
df_actualizado
```

```python
# SIMULACION MONTECARLO

import numpy as np
import pandas as pd

num_simulations = 1000
products = df_actualizado['ProductName'].unique()
sales_results = {product: [] for product in products}

for _ in range(num_simulations):
    numero_aleatorio = np.random.uniform(0, df_actualizado['CumulativeProbability'].max())

    for index, row in df_actualizado.iterrows():
        if numero_aleatorio <= row['CumulativeProbability']:
            sales_results[row['ProductName']].append(row['Quantity'])
            break

# Determinar el valor de compra más frecuentemente registrado para cada producto
most_frequent_sales = {}

for product, quantities in sales_results.items():
    if quantities:  # Verificar que haya ventas registradas
        most_frequent = pd.Series(quantities).value_counts().idxmax()
        most_frequent_sales[product] = most_frequent
    else:
        most_frequent_sales[product] = 0  # No hubo ventas en las simulaciones para este producto

# Mostrar los resultados con formato alineado
max_length = max(len(product) for product in most_frequent_sales.keys())
print(f"{'Producto'.ljust(max_length)} | Número más frecuente de ventas en {num_simulations} simulaciones")
print('-' * (max_length + 54))

for product, frequent_sales in most_frequent_sales.items():
    print(f"{product.ljust(max_length)} | {frequent_sales}")
```

```python
# SIMULACION CONSIDERANDO PROBABILIDAD DE VENTA DEL PRODUCTO

num_simulations = 1000
results = {}

for product in df_actualizado['ProductName'].unique():
    counts = []
    product_data = df_actualizado[df_actualizado['ProductName'] == product]
    sales_probability = product_data['SalesProbPerProd'].iloc[0]

    for _ in range(num_simulations):
        # Simular el número de ventas por vuelo, considerando la probabilidad de venta del producto
        sales = np.random.binomial(n=int(product_data['Passengers'].iloc[0]), p=sales_probability)
        counts.append(sales)

    # Encontrar el número más frecuente de ventas
    most_frequent_sales = pd.Series(counts).value_counts().idxmax()
    results[product] = most_frequent_sales

# Mostrar los resultados con formato alineado
max_length = max(len(product) for product in results.keys())
print(f"{'Producto'.ljust(max_length)} | Número más frecuente de ventas en {num_simulations} simulaciones")
print('-' * (max_length + 54))

for product, frequent_sales in results.items():
    print(f"{product.ljust(max_length)} | {frequent_sales}")
```
